```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Data/wrds_adj.csv", engine="pyarrow",
                   parse_dates=True) 
data.drop("PERMCO", axis=1, inplace=True)
data
```

```python
stock = pd.read_pickle("Data/allstock_reduced.pickle")
col = stock.columns
```

```python
sp500 = data.loc[data.PERMNO.isin(col)]
sp500
```

```python
sp500.loc[:,"PRC"] = np.abs(sp500.PRC)
```

```python
sp500_drop = sp500.dropna()
```

```python
sp500_drop.loc[:,"Adj_close"] = sp500_drop.PRC / sp500_drop.CFACPR
```

```python
data_pp = sp500_drop.reset_index(drop=True)[["PERMNO","date","Adj_close"]]
data_pp
```

```python
data_pp.to_pickle("Data/allstock_adj.pickle")
```
